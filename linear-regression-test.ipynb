```python
#setup
import numpy as np
import pandas as pd
import random
from matplotlib import pyplot as plt
```

```python
#randomly generate weight and bias
w=random.randrange(-1000000,1000000)
b=random.randrange(-1000000,1000000)
print(w,b)
def func(x):
    return (w*x+b)
```

```python
#house data
house_data_volume=np.array([0,90,63.77,92.97,105.33,103,52,38])
house_data_price=np.array([0,4500000,3850000,6000000,6100000,4880000,1650000,1180000])
max_volume=0
max_price=0
for i in house_data_volume:
    max_volume=max(max_volume,i)
for i in house_data_price:
    max_price=max(max_price,i)
print(max_volume,max_price)
```

```python
#cost
def cost():
    sum=0
    m=house_data_volume.size
    for i in range (m):
        sum+=(func(house_data_volume[i])-house_data_price[i])*(func(house_data_volume[i])-house_data_price[i])
    return sum/m
```

```python
num=max(max_volume,(max_price-b)/w)
plt.scatter(house_data_volume,house_data_price)

alpha=0.00015
cnt=10
data_cost_x=np.zeros(cnt)
data_cost_y=np.zeros(cnt)

data_w_x=np.zeros(cnt)
data_w_y=np.zeros(cnt)

data_b_x=np.zeros(cnt)
data_b_y=np.zeros(cnt)
for j in range (cnt):
    plt.plot([0,num],[func(0),func(num)],color=(0,0,min(1,j/10)))
    data_cost_x[j]=j
    data_w_x[j]=j
    data_b_x[j]=j
    sum=0
    sum2=0
    m=house_data_volume.size
    for i in range (m):
        sum+=(func(house_data_volume[i])-house_data_price[i])*(house_data_volume[i])
        sum2+=(func(house_data_volume[i])-house_data_price[i])
    sum/=m
    sum2/=m
    w=w-alpha*sum
    b=b-alpha*sum2
    data_cost_y[j]=cost()
    data_w_y[j]=w
    data_b_y[j]=b
#     print(round(w,2),round(b,2))
print(cost())
print(w,b)

#10000 train
# 536343786850.1025
# 54731.95159813268 -127971.4518575691

#20000 train
# 503968626651.8245
# 56550.714653318195 -283666.97219868656

#30000 train
# 479283189733.48236
# 58138.86118283692 -419620.5256790963

#40000 train
# 460461018905.06305
# 59525.63302869734 -538335.3651520292

#50000 train
# 445694900261.78625
# 60774.16913960915 -645216.5110488824

#...

#100000 train
#439172043802.88116
#59149.22751319445 -506102.2622790677
```
